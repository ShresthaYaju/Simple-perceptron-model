```python
import numpy as np
import pandas as pd


def activation_function(sum):
    if sum >= 1:
        return 1
    else:
        return 0


def OR_gate(input1, input2, target_output, learning_rate):
  
    input_array = np.array([[input1[i], input2[i]] for i in range(len(input1))])


    weight_array = np.array([0.5, 0.5])


    bias = -1

    num_iterations = 5


    iteration_data = []


    for iteration in range(num_iterations):

        for i in range(len(target_output)):

            sum = np.dot(input_array[i], weight_array) + bias


            output = activation_function(sum)


            error = target_output[i] - output


            weight_correction = learning_rate * error * input_array[i]


            weight_array += weight_correction

            
            iteration_data.append([iteration, input_array[i][0], input_array[i][1], weight_array[0], weight_array[1], bias, sum, output, target_output[i], error, weight_correction[0], weight_correction[1]])


    column_names = ['Iteration', 'Input 1', 'Input 2', 'Weight 1', 'Weight 2', 'Bias', 'Sum', 'Output', 'Target Output', 'Error', 'Weight Correction 1', 'Weight Correction 2']
    iteration_df = pd.DataFrame(iteration_data, columns=column_names)


    print("Iteration Data:")
    print(iteration_df)


    print("OR Gate Truth Table:")
    print("Input 1 | Input 2 | Output")
    print("---------------------------")
    for i in range(2):
        for j in range(2):

            sum = np.dot(np.array([i, j]), weight_array) + bias

            output = activation_function(sum)

       
            print(f"   {i:^6}  |  {j:^6}  |  {output:^6}")
    print("---------------------------")
    iteration_df.to_csv('iteration_data.csv')


input1 = np.array([0, 0, 1, 1])
input2 = np.array([0, 1, 0, 1])
target_output = np.array([0, 1, 1, 1])
learning_rate = 0.1

OR_gate(input1, input2, target_output, learning_rate)
```
